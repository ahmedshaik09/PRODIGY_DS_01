```python
import matplotlib.pyplot as plt

# Sample data (replace this with your data)
categories = ['Male', 'Female', 'Non-Binary', 'Other']
counts = [300, 450, 75, 25]

# Create the bar chart
plt.bar(categories, counts)

# Adding labels and a title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution in a Population')

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample age data (replace this with your data)
ages = [28, 32, 45, 50, 18, 60, 35, 28, 30, 22, 42, 55, 26, 29, 31]

# Create the histogram
plt.hist(ages, bins=10, edgecolor='k')  # You can adjust the number of bins as needed

# Adding labels and a title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in a Population')

# Show the histogram
plt.show()

```
